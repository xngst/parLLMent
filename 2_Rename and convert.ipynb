### A parlementi napló fájlok átnevezése kiadási számok alapján

```python
import os

from orszaggyulesi_naplo import hunparl as hp
from pathlib import Path
from datetime import datetime 

in_dir = Path("parllment/raw_pdf/2014-2018")
out_dir = Path("parllment/txt/2014-2018")

def convert_hungarian_date(date_str):
    honapok = {
        'január': '1', 
        'február': '2', 
        'március': '3', 
        'április': '4',
        'május': '5', 
        'június': '6',
        'július': '7', 
        'augusztus': '8',
        'szeptember': '9', 
        'október': '10', 
        'november': '11', 
        'december': '12'
    }
    parts = date_str.replace('.', '').split()
    year = parts[0]
    month = honapok[parts[1].lower()]
    day = parts[2]
    dt = datetime(int(year), int(month), int(day))
    return dt.strftime('%Y-%m-%d')
```

```python
for file in os.listdir(in_dir):
    file_path = Path(in_dir/file)
    naplo = hp.OgyNaplo(file_path)
    datum = convert_hungarian_date(naplo.datum)
    ciklus = naplo.ciklus.split(".")[0]
    szam = naplo.szam
    try:
        new_file_name = in_dir/f"{ciklus}_{datum}_{szam}.pdf"
        print(new_file_name)
        file_path.rename(new_file_name)
    except Exception as e:
        print(file_path)
        print(e)
        continue
```

### utf8 kinyerése PDF-ből

A kód végigiterál egy mappányi országgyűlési naplón, és egy speciális feldolgozóegység segítségével kinyeri belőlük a már megtisztított szöveges tartalmat. A folyamat során a PDF-fájlok nevét szöveges kiterjesztésre cseréli, ügyelve a magyar ékezetes karakterek megőrzésére az UTF-8 kódolással. Az eredményül kapott .txt fájlokat egy külön kimeneti könyvtárba menti el, előkészítve az adatokat a későbbi elemzéshez vagy adatbázisba építéshez.

```python
for file in os.listdir(in_dir):
    file_path = Path(in_dir/file)
    naplo = hp.OgyNaplo(file_path)
    text = naplo.tisztazott
    file = file.replace("pdf","txt")
    with open(out_dir/file,"w",encoding="utf-8") as f:
        f.writelines(text)
```
